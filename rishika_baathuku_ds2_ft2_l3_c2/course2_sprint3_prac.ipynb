```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

## Tanh activation function  
     It returns the value (1-exp(-2x))/(1+exp(-2x)) and the value returned will be lies in between
 -1 to 1.

```python
def tanh(x):
    return np.tanh(x)
x = np.linspace(-10, 10)
plt.plot(x, tanh(x))
plt.axis('tight')
plt.title('Activation Function :Tanh')
plt.show()
```

## Sigmoid activation function  
    It returns 1/(1+exp(-x)). where the values lies between zero and one 

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

## Softmax activation function
   Compute softmax values for each sets of scores in x.
Softmax turns logits, the numeric output of the last linear layer
of a multi-class classification neural network into probabilities.

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x), axis=0)
x = np.linspace(-10, 10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation Function :Softmax')
plt.show()
```

```python
# Generate some 130 training data between -15 to 15.
sample=np.random.randint(-15,15,130)
sample
```

```python
import pandas as pd
data=pd.DataFrame(sample)
data.reset_index(level=0, inplace=True)
data
```

```python
# Generate label data using any one activation function defined in the Task â€“ 1
labels=np.tanh(data)
labels
```

```python
# Plot input data
plt.figure()
plt.hist(data[0])
plt.title('Histogram')
plt.show()
```

```python
#Define minimum and maximum values for each dimension
dim1_min, dim1_max, dim2_min, dim2_max = 0, 1, 0, 1
```

```python
# Set different layers of perceptron
# Number of neurons in the output layer
num_output = labels.shape[1]
num_output
```

```python
#Define a perceptron with 2 input neurons (because we have 2dimensions in the input data)
import neurolab as nl
dim1 = [dim1_min, dim1_max]
dim2 = [dim2_min, dim2_max]
perceptron = nl.net.newp([dim1, dim2], num_output)
perceptron
```

```python
# Train the perceptron using the data
error_progress = perceptron.train(data, labels, epochs=100, show=20, lr=0.03)
```

```python
# Plot the training progress
plt.figure()
plt.plot(error_progress)
plt.xlabel('Number of epochs')
plt.ylabel('Training error')
plt.title('Training error progress')
plt.grid()
plt.show()
```
