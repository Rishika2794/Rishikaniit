# Task1 (Extracting Western Subsection from URL)

```python
from pattern.web import URL, plaintext

url0='https://en.wikipedia.org/wiki/Emoticon'
print(url0)
byt = URL(url0).download() # byte object
s = str(byt,'utf-8')       # byte to string
s = plaintext(s, keep={'h1':[], 'h2':[], 'h3':[]})
print(s)
```

```python
import re
```

```python
section_name = "Styles"
subsection_name = "Western"
```

```python
header2_pattern = re.compile("<h2>\s*(.*)\s*\[edit\]\s*</h2>")
```

```python
h2s = header2_pattern.finditer(s)
```

```python
style_section_match = None
next_section_match = None
for h2 in h2s:
    if h2.group(1) == "Styles":
        style_section_match = h2
    elif style_section_match is not None and next_section_match is None:
        next_section_match = h2
```

```python
style_section_text = ""
if next_section_match is None:
    style_section_text = s[style_section_match.end():]
else:
    style_section_text = s[style_section_match.end():next_section_match.start()]
```

```python
len(style_section_text)
```

```python
header3_pattern = re.compile("<h3>\s*(.*)\s*\[edit\]\s*</h3>")
```

```python
h3s = header3_pattern.finditer(style_section_text)
```

```python
western_section_match = None
next_section_match = None
for h3 in h3s:
    if h3.group(1) == "Western":
        western_section_match = h3
    elif western_section_match is not None and next_section_match is None:
        next_section_match = h3
```

```python
western_section_text = ""
if next_section_match is None:
    western_section_text = s[style_section_match.end()+western_section_match.end():]
else:
    western_section_text = s[style_section_match.end()+western_section_match.end():style_section_match.end()+next_section_match.start()]
```

```python
western_section_text
```

```python
print('Length of subsection "Western":', len(western_section_text))
```

```python
re.compile("[.!?:;]").split(western_section_text)
```

```python
import nltk
from nltk.tokenize import sent_tokenize
nltk.download("punkt")
```

```python
sentences = sent_tokenize(western_section_text)
```

```python
whitespace_tokens = []
for sentence in sentences:
    whitespace_tokens.append(re.compile("\s").split(sentence))
```

```python
whitespace_tokens
```

# Task 2 (Different types of Tokenizers)

```python
from nltk.tokenize import SpaceTokenizer
from nltk.tokenize import TweetTokenizer
from nltk.tokenize import word_tokenize
from nltk.tokenize import TreebankWordTokenizer
```

```python
word_tokenizer_tokens = []
for sentence in sentences:
    word_tokenizer_tokens.append(word_tokenize(sentence))
```

```python
word_tokenizer_tokens
```

```python
st = SpaceTokenizer()
```

```python
st_tokenizer_tokens = []
for sentence in sentences:
    st_tokenizer_tokens.append(st.tokenize(sentence))
```

```python
st_tokenizer_tokens
```

```python
tt = TweetTokenizer()
```

```python
tt_tokenizer_tokens = []
for sentence in sentences:
    tt_tokenizer_tokens.append(tt.tokenize(sentence))
```

```python
tt_tokenizer_tokens
```

```python
tree = TreebankWordTokenizer()
```

```python
tree_tokenizer_tokens = []
for sentence in sentences:
    tree_tokenizer_tokens.append(tree.tokenize(sentence))
```

```python
tree_tokenizer_tokens
```

# Task 3 (Count Vectorizer)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvec = CountVectorizer(min_df=1, tokenizer=tt.tokenize)

# sents turned into sparse vector of word frequency counts
sents_counts = cvec.fit_transform(sentences)
# foovec now contains vocab dictionary which maps unique words to indexes
print('vocabulary mapped to numbers\n',cvec.vocabulary_)
```

```python
print('word counts for each sentence\n',sents_counts.toarray())
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tfidf_transformer = TfidfTransformer()
sents_tfidf = tfidf_transformer.fit_transform(sents_counts)
print('word tfidf scores for each sentence\n',sents_tfidf.toarray())
```
